```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def gauss(x, A, sigma):
    return A * np.exp(-(x ** 2) / (sigma ** 2))
```

```python
def own_mean(x, i):
    r = 0
    imin = i - 2
    imax = i + 2
    for j in range(imin, imax + 1, 1):
        if j >= 0 and j < len(x):
            r = r + x[j]
    r = r / 5.0
    return r
```

```python
def own_med(x, i):
    imin = i - 1
    imax = i + 1
    ir = 0
    if imin < 0:
        ir = x[imax]
    else:
        if imax >= len(x):
            ir = x[imin]
        else:
            arr = [x[imin], x[i], x[imax]]
            arr.sort()
            ir = np.median(arr)
    return ir
```

```python
F = 3
dt = 0.05

x = np.arange(-F, F, dt)

yx = gauss(x, 1, 1)
uxbase = gauss(x, 1, 1)
ux = gauss(x, 1, 1)
N = len(yx)
```

```python
a = 0.25
epsv = 0.05
```

```python
px = a * np.random.rand(1, 20)[0]
```

```python
pos = np.array([25, 35, 40, 54, 67, 75, 95])
pxx = len(pos)
```

```python
for i in range(0, pxx, 1):
    ux[pos[i]] = ux[pos[i]] + px[i]
    uxbase[pos[i]] = uxbase[pos[i]] + px[i]
```

```python
for i in range(0, N, 1):
    smthm = own_mean(ux, i)
    if abs(ux[i] - smthm) > epsv:
        ux[i] = smthm
```

```python
figure, ax = plt.subplots(figsize=(8, 8))
ax.grid(linewidth=0.5)
plt.title('Mean')
plt.plot(x, uxbase, color='r', label='Initial')
plt.plot(x, ux, color='g', label='Fixed')
plt.legend()
```
